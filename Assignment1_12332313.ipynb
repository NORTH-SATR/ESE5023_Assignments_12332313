## PS1_3 Pascal triangle

```python
#question 3 
#I got inspired by reading 
#ways:https://www.mathsisfun.com/pascals-triangle.html 
#https://blog.csdn.net/m0_62338174/article/details/129746943 
 
def Pascal_triangle(k): 
    row = np.zeros(k+1, dtype=float) 
    row[0] = 1 
    for i in range(1, k + 1): 
        row[i] = row[i - 1] * (k - i + 1) /i  
    return row 
print("Pascal triangle（k = 100) :") 
print(Pascal_triangle(99)) 
print("Pascal triangle（k = 200) :") 
print(Pascal_triangle(199)) 
```

## PS1_4 Add or double

```python
#question 4 
 
def Least_moves(x): 
    steps = np.full(x + 1, 0) 
    for i in range(2, x + 1): 
        if i % 2 == 0: 
            steps[i] = steps[i // 2] + 1 
        else: 
            steps[i] = steps[i - 1] + 1 
    print(int(steps[x])) 
Least_moves(2)   
Least_moves(5)  
```

```python
def find_num(item, count_list): # 用于找出目标数字的在list中的位置
    idx_list = []
    for index, element in enumerate(count_list):
        if element == item:
            idx_list.append(index+1)
#             print(index+1, end=" ")
    return idx_list

# 找到表达式最少的数字
num_minexpr = find_num(min(Total_solutions),Total_solutions)
print("num with Min exprs: "+str(num_minexpr)+", the amount of exprs: "+str(min(Total_solutions)))

# 找到表达式最多的数字
num_maxexpr = find_num(max(Total_solutions),Total_solutions)
print("num with Max exprs: "+str(num_maxexpr)+", the amount of exprs: "+str(max(Total_solutions)))
```
